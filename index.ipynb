## Employees and their Office (a One-to-One join)

Return a list of all of the employees with their first name, last name and the city and state of the office that they work out of (if they have one). Include all employees and order them by their first name, then their last name.

```python
#Your code here
import sqlite3
conn = sqlite3.connect('data.sqlite')
cursor = conn.cursor()



import pandas as pd
#Your code here
def query_to_df(query):
    df = pd.DataFrame(cursor.execute(query).fetchall())
    df.columns = [i[0] for i in cursor.description]
    return df


query_to_df("""select firstName,  lastName, city, state from employees join offices order by firstName,  lastName""")
```

```python
query_to_df("""select * from offices""")
```

## Customers and their Orders (a One-to-Many join)

Return a list of all of the customers. For each customer return a record for each of their order numbers, order dates and statuses.

```python
# Your code here
query_to_df("""select customerName, orderNumber, orderDate, status from customers join orders""")
```

## Orders and their Order Details (another One-to-Many join)

Return a list of orders. For each order return a record for each order detail within the order.

```python
# Your code here
query_to_df("""select orders.orderNumber, productCode, quantityOrdered,priceEach, orderLineNumber from orders join orderdetails """ )
```

## Orders, Order details and Product Details (a Many-to-Many Join)

Return a list of the orders with the the order information, and all the product information for each product in the order. 

```python
# Your code here

query_to_df(""" select orderNumber, productName, quantityOrdered, buyPrice, sum( quantityOrdered*buyPrice) as Total from orders join orderdetails using(orderNumber) join products using(productCode) group by orderNumber""")

```

```python
query_to_df(""" select orderNumber, productName, quantityOrdered, buyPrice,  quantityOrdered*buyPrice as Total from orders join orderdetails using(orderNumber) join products using(productCode) """)

```
